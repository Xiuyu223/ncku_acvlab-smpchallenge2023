```python
import os, pdb
from shutil import copyfile

with open('data_2022/train/train_img.txt', 'r') as fp:
    data=fp.readlines()

with open('data_2022/train/train_category.txt', 'r') as fp:
    cat=fp.readlines()
len1=len(data)
fp=open('data_2022/train/train_img_list.txt','w')
fp3=open('data_2022/train/train_img_all_list.txt','w')

fp2=open('data_2022/train/train_img_exp_list.txt','w')

fp.write('Uid Pid Path\n')
fp2.write('Uid Pid Path\n')

cnt = 0
print('Total number of files:', len1)
for i in range(len1):
    line=data[i] 
    line2=cat[i+1]
    _,_,user,photo=line[8:].split('/') # For train
#     _,user,photo=line.split('/') # For test
    photo=photo.replace('\n','')
    uid,pid,_,_,_=line2.split(' ')
    pdb.set_trace()
    if os.path.isdir('imgs/%s/%s/' % (user,photo)): # Put the training image in dir imgs
        dx=os.listdir('imgs/%s/%s/' % (user,photo))
        
        if len(dx)>0:
            pth='imgs/%s/%s/%s' % (user,photo,dx[0])
            fp.write('%s %s %s\n' % (uid,pid,pth))
            fp3.write('%s %s %s\n' % (uid,pid,pth))
        else:
#             print('Not found',pth)
            cnt+=1
            fp2.write('%s %s %s\n' % (uid,pid, "None"))
            fp3.write('%s %s %s\n' % (uid,pid, "None"))
    else:
#         print('Not found',pth)
        cnt+=1
        fp2.write('%s %s %s\n' % (uid,pid, "None"))
        fp3.write('%s %s %s\n' % (uid,pid, "None"))
fp.close()    
fp2.close()
fp3.close()
print('Not found files ', cnt)
```

```python
import os, pdb
from shutil import copyfile

with open('data_2022/test/test_img.txt', 'r') as fp:
    data=fp.readlines()

with open('data_2022/test/test_category.txt', 'r') as fp:
    cat=fp.readlines()
    
print("Length for img and cat are %d and %d" % (len(data), len(cat)))
len1=len(data)
fp=open('data_2022/test/test_img_list.txt','w')
fp3=open('data_2022/test/test_img_all_list.txt','w')

fp2=open('data_2022/test/test_img_exp_list.txt','w')

fp.write('Uid Pid Path\n')
fp2.write('Uid Pid Path\n')

cnt = 0
print('Total number of files:', len1)
for i in range(len1):
    line=data[i] 
    line2=cat[i+1]
    _,user,photo=line.split('/') # For test
    photo=photo.replace('\n','')
    uid,pid,_,_,_=line2.split(' ')
    if os.path.exists('test/imgs/%s/%s' % (user,photo)):
        
        pth='test/imgs/%s/%s' % (user,photo)
        fp.write('%s %s %s\n' % (uid,pid,pth))
        fp3.write('%s %s %s\n' % (uid,pid,pth))

    else:
#         print('Not found',pth)
        cnt+=1
        fp2.write('%s %s %s\n' % (uid,pid, "None"))
        fp3.write('%s %s %s\n' % (uid,pid, "None"))
fp.close()    
fp2.close()
fp3.close()
print('Not found files ', cnt)
```
