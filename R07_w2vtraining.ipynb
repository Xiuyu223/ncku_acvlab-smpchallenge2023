```python
import gc
gc.collect()

import logging

from gensim.models import word2vec



logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = word2vec.LineSentence("data_2022/train/train_cat_pure_text.txt")
model =word2vec.Word2Vec(sentences, vector_size=20 ,window=6, workers=10, min_count=0, epochs=10)
model.save("word2vec_cat.model")

sentences2 = word2vec.LineSentence("data_2022/train/train_tags_pure_text.txt")
model2 =word2vec.Word2Vec(sentences2, vector_size=20 ,window=7, workers=10, min_count=10, epochs=10)
model2.save("word2vec_tag.model")

sentences3 = word2vec.LineSentence("data_2022/train/train_caption_only.txt")
model3 =word2vec.Word2Vec(sentences3, vector_size=20 ,window=20, workers=10, min_count=5, epochs=10)
model3.save("word2vec_cap.model")

```
