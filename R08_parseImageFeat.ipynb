```python
##Please use TF_Feature_Exatractor to get the image features
##https://github.com/tomrunia/TF_FeatureExtraction


import pandas as pd

df_imfeat = pd.read_csv('data_2022/test/test_image_feat.txt', delimiter=',')
df_imlist = pd.read_csv('data_2022/test/test_img_all_list.txt', delimiter=' ')
```

```python
df_imlist
```

```python
import numpy as np
import pdb

x1 = df_imfeat['Uid'].values + np.asarray(df_imfeat['Pid'].values, np.str)
y1 = df_imlist['Uid'].values + np.asarray(df_imlist['Pid'].values, np.str)

len1=y1.shape[0]
feat = []
for i in range(len1):
    ids = y1[i]
    try:
        idx = np.where(x1==ids)[0][0]
        feat.append(df_imfeat['Feat'].values[idx])
    except:
        zs = ""
        for k in range(1001):
            zs = zs + " 0"
        zs += '\n'
        feat.append(zs)
        
```

```python
def createlist(ls_name,number):
    ls=list()
    for i in range(0,number):
        name=ls_name +'_'+ str(i+1)
        ls.append(name)
    return ls

len(feat)
df_imlist = df_imlist.drop(['Path'], axis=1)
feat = pd.DataFrame(feat)
feat.columns  = createlist('feat',feat.values.shape[1])
df_imlist = df_imlist.join(feat)
df_imlist.to_csv('data_2022/test/test_image_feat2.csv')
```
